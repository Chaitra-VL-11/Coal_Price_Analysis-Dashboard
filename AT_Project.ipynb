####**Measure of Central Tendency**

1) **Mean**

```python
Mean=df.mean(numeric_only=True)
```

```python
print("Mean for each column:")
print(Mean)
```

2) **Median**

```python
Median=df.median(numeric_only=True)
```

```python
print("Median for each column:")
print(Median)
```

3) **Mode**

```python
Mode=df.mode().iloc[0]
```

```python
print("Mode for each column:")
print(Mode)
```

####**Measure of Dispersion**

1) **Range**

```python
Range=df.select_dtypes(include=['number']).apply(lambda x: x.max()-x.min())
print("Range for each column :")
print(Range)
```

2) **Standard Deviations**

```python
std=df.std(numeric_only=True)
print("Standard Deviation for each column:")
print(std)
```

3) **Variance**

```python
Variance=df.var(numeric_only=True)
print("Variance for each column:")
print(Variance)
```

####**Skewness** : Skewness can have a value that is either positive, Negative, or even undefined. If the distribution is not symmetrical, the box and whisker plot is said to be a skewed box plot.

```python
Skewness=df.skew(numeric_only=True)
print("Skewness for each column:")
print(Skewness)
```

```python
from scipy.stats import skew
float_column=df.select_dtypes(include=['float']).columns
skewness_values=df[float_column].apply(lambda x: skew(x, nan_policy='omit'))
plt.figure(figsize=(15,14))
for i, col in enumerate(float_column):
  plt.subplot(6,3,i+1)
  sns.boxplot(data=df,x=col,color='skyblue')
  plt.title(f'{col}-Skewness:{skewness_values[col]:.2f}')
plt.tight_layout()
plt.show()
```

####**Kurtosis** : Kurtosis is a measure of the tailedness of a distribution.

```python
Kurtosis=df.kurt(numeric_only=True)
print("Kurtosis for each column:")
print(Kurtosis)
```

A standard normal distribution has kurtosis of 3 and is recognized as mesokurtic.

```python
from scipy.stats import kurtosis
float_column=df.select_dtypes(include=['float','int']).columns
kurtosis_values=df[float_column].apply(lambda x: kurtosis(x,fisher=True, nan_policy='omit'))
print("kurtosis values for numeric columns:")
print(kurtosis_values)
plt.figure(figsize=(9,6))
sns.barplot(x=kurtosis_values.index,y=kurtosis_values.values,palette='rainbow')
plt.axhline(0,color='red',linestyle='--',label='Normal Distribution(kurtosis=0)')
plt.title('Kurtosis of numeric columns')
plt.ylabel('kurtosis values')
plt.xlabel('Columns')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```
